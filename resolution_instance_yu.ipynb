```python
import numpy as np
```

```python
#M :ligne
class Robot_resolution:
    #========== initialisation ==================
    def __init__(self,path):
        file=open(path,"r")
        self.ID_ORIENTATIONS = {d: i for i,d in enumerate(["nord","est","sud","ouest"])}
        self.M,self.N=-1,-1
        self.D1,self.D2,self.F1,self.F2,self.direction=-1,-1,-1,-1,None
        self.matrice=[]
        self.lire_fichier(file)

    #========== lecture fichier ===============
    def lire_fichier(self,file):
        """lecture fichier d'entrée"""
        tmpm=0
        for n,line in enumerate(file.readlines()):
            tmp=line.replace("\n","").split(" ")
            if self.M==-1:
                #taille de matrice
                vals=[int(n) for n in tmp]
                self.M,self.N=vals[0],vals[1]
            elif tmpm<self.M:
                #matrice
                vals=list()
                assert len(tmp)==self.N, f"fichier erreur: nb données incorrecte à la ligne {n}"
                for i in range(self.N):
                    try:
                        vals.append(int(tmp[i]))
                    except:
                        raise ValueError
                tmpm+=1
                self.matrice.append(vals)
                self.matrice=np.array(self.matrice)
            else:
                #coordonnées de départ et de destination avec le direction initial de robot
                if self.D1==-1:
                    assert len(tmp)==5, "fichier erreur: nombre de paramètres incorrecte"
                    try:
                        self.D1,self.D2,self.F1,self.F2=(int(tmp[i]) for i in range(4))
                        self.direction=tmp[4]
                    except:
                        raise ValueError
    def test_init(self):
        """affiche tous les variables"""
        print(f"(D1,D2,F1,F2)=({self.D1},{self.D2},{self.F1},{self.F2})\
              \ndirection: {self.direction}\nmatrice: {self.matrice}")
        
    #============ traitement ===================
    def tourne(self,o:str="gauche"):
        idx=(self.ID_ORIENTATIONS[o]+1)%4 if o=="gauche" else (self.ID_ORIENTATIONS[o]-1)%4
        # ou self.direction = self.ID_ORIENTATIONS[idx], tous dépendent comment on va faire
        return self.ID_ORIENTATIONS[idx]
```

```python
r=Robot_resolution("instances.txt")
r.test_init()
```
